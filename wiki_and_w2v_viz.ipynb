```python
import os
import re
```

```python
os.listdir('text')
```

```python
def list_wiki(dirname):
    filepaths = []
    filenames = os.listdir(dirname)
    for filename in filenames:
        filepath = os.path.join(dirname, filename)

        if os.path.isdir(filepath):
            # 재귀 함수
            filepaths.extend(list_wiki(filepath))
        else:
            find = re.findall(r"wiki_[0-9][0-9]", filepath)
            if 0 < len(find):
                filepaths.append(filepath)
    return sorted(filepaths)

filepaths = list_wiki('text')
```

```python
len(filepaths)
```

```python
with open("data/wiki_output_file.txt", "w") as outfile:
    for filename in filepaths:
        with open(filename) as infile:
            contents = infile.read()
            outfile.write(contents)
```

# W2V visualization

```python
with open('data/euphemism_answer_drug.txt') as euph:
    line = euph.readlines()

    euph_dict = {}
    for i in line:
        key = i.strip().split(':')[0]
        values = i.strip().split(':')[1]
        value = values.strip().split('; ')
        euph_dict[key.lower()] = [val.lower() for val in value]
```

```python
# k=10
with open('data/output/w2v/wiki_eval_w2v_embed_top10.txt') as fn:
    line = fn.readlines()
    result_dict = {}
    for i in line[1:]:
        l = i.strip().split(', ')
        key = l[0]
        value = l[1:]
        result_dict[key] = value

denom = len(key) * len(value)
count_correct = 0
for key in result_dict:
    if key in euph_dict:
        for val in result_dict[key]:
            if val in euph_dict[key]:
                print(f"formal drug name: {key} <-> euphemistic word: {val}")
                count_correct += 1

print(f"\ncorrect words: {count_correct} & top-10 precision: {count_correct / denom}")
```
